#### Feature Engineering

```python
final_df['Date_of_Journey'] = pd.to_datetime(final_df['Date_of_Journey'], format="%d/%m/%Y")

final_df['Date'] = final_df['Date_of_Journey'].dt.day
final_df['Month'] = final_df['Date_of_Journey'].dt.month
final_df['Year'] = final_df['Date_of_Journey'].dt.year

final_df.drop('Date_of_Journey',axis=1, inplace=True)
```

Note:- When you'll do feature engineering try to catch similar types of fields first

```python
# Feature Engineering on Arrival_Time
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
final_df.head()
```

```python
final_df['Arrival_hour'] = final_df.Arrival_Time.apply(lambda x:x.split(':')[0])
final_df.head(1)
```

```python
final_df['Arrival_min'] = final_df.Arrival_Time.apply(lambda x:x.split(':')[1])
final_df.head(1)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df.info()
```

```python
# Coverting these two features into int datatype
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df.head(1)
```

```python
### Converting Departure Time into two features Dep_hour and Dep_min

final_df['Dep_hour'] = final_df['Dep_Time'].apply(lambda x:x.split(':')[0])
final_df['Dep_min'] = final_df['Dep_Time'].apply(lambda x:x.split(':')[1])

## Converting object to int datatype
final_df['Dep_hour'] = final_df['Dep_hour'].astype(int)
final_df['Dep_min'] = final_df['Dep_min'].astype(int)

## Deleting the old feature

final_df.drop('Dep_Time',axis=1,inplace=True)


## Looking Dataset
final_df.head(2)
```

```python
final_df['Total_Stops'].isnull().sum()
```

```python
# final_df['Total_Stops'] = final_df['Total_Stops'].fillna()
```

```python
final_df.Total_Stops.unique()
```

```python
final_df.Total_Stops = final_df.Total_Stops.map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
# Deleting Route Column
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df
```

```python
final_df['Additional_Info'].unique()
```

```python
# Extract hour part (returns 0 if 'h' is not in the string)
final_df['Duration_hour'] = final_df['Duration'].apply(
    lambda x: int(x.split('h')[0].strip()) if 'h' in x else 0
)

final_df['Duration_min'] = final_df['Duration'].apply(
    lambda x: int(x.split('h')[1].replace('m', '').strip()) if 'h' in x and 'm' in x
    else int(x.replace('m', '').strip()) if 'm' in x else 0
)

final_df['Duration_total_min'] = final_df['Duration_hour'] * 60 + final_df['Duration_min']
```

```python
final_df.drop('Duration',axis=1, inplace=True)
final_df.drop('Duration_hour',axis=1, inplace=True)
final_df.drop('Duration_min',axis=1, inplace=True)
final_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
final_df['Airline'] = encoder.fit_transform(final_df['Airline'])
final_df['Source'] = encoder.fit_transform(final_df['Source'])
final_df['Destination'] = encoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = encoder.fit_transform(final_df['Additional_Info'])
```
