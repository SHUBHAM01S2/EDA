```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# Combine the dataframes 
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
# Observations: In india the main base of Zomato
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
## Pie Chart- Top 3 Countries
plt.pie(country_values[:3], labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

# Observation: The main base of Zomato is in India
---
1. The main Market of Zomato is in India
2. Zomato maximum records or transactions are from India After that USA and then UK


```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# Observation: Rating of Zomato.
---
1. When Rating is between 4.5 to 4.9 ---> Excellent
2. When Ratings are between 4.0 to 4.4 ---> Very good
3. When Rating is between 3.5 to 3.9 ---> good
4. When Rating is between 3.0 to 3.4 ---> average
5. When Rating is between 2.5 to 2.9 ---> average
6. When Rating is between 1.8 to 2.4 ---> poor


```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = "Aggregate rating", y="Rating Count", data=ratings)
plt.show()
```

```python
import matplotlib
sns.barplot(x = "Aggregate rating", y="Rating Count",hue='Rating color', data=ratings, palette=['black','Red','Orange','Yellow','Green','Green'])
plt.show()
```

## Observations
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
# Count Plot
sns.countplot(x = 'Rating color', data=ratings, palette=['black','red','orange','yellow','green','green'])
plt.show()
```
