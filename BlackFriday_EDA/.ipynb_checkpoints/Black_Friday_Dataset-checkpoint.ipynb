## Cleaning and preparing the data for model training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# importing the dataset
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
# Importing the test dataset
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
### Merge both train and test data
df = pd.concat([df_train, df_test], ignore_index=True)
df.head()
```

```python
## Basic 
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1, inplace=True) # inplace = True means that it reflect the changes in the main dataset
```

```python
df.head()
```

```python
## Data Preprocessing
pd.get_dummies(df['Gender'])
```

```python
## Handling the categorical feature Gender
df['Gender'] = df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
## Handle Categorical Feature Age
df['Age'].unique() #=> gives how many values are there
```

```python
df['Age'] = df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6, '55+':7})
```

```python
## 2nd Technique

# from sklearn.preprocessing import LabelEncoder

# # Label_encoder object knows how to understand word labels.
# label_encoder = LabelEncoder()

# # Encode labels in cloumn 'Age
# df['Age'] = label_encoder.fit_transform(df['Age'])
# df['Age'].unique()
```
